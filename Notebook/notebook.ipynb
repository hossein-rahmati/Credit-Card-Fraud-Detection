### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Dataset

```python
PATH = str("../Data/creditcard.csv")
ds = pd.read_csv(PATH)
ds.head()
```

```python
ds.describe()
```

Cheking if dataset has missing values

```python
ds.isnull()
```

Seprate dataset to fraud and none fraud transactions

```python
fraud = ds[ds["Class"] == 1]
valid = ds[ds["Class"] == 0]

print("Valid transaction: ", len(valid))
print("Fraud transaction: ", len(fraud))
print((len(fraud) / len(valid)) * 100)
```

About 0.17% of this dataset includes fraud transactions, so the dataset is crucialy imbalanced

### Correlation matrix

```python
correlation_matrix = ds.corr()
sns.heatmap(correlation_matrix, vmax=0.8, square=True)
```

### Data preprocessing

```python
X = ds.iloc[:, :-1].values
y = ds.iloc[:, -1].values

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=32
)
```

### Training Random Forest Classifier Model on the training set

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
yPred = classifier.predict(X_test)
```
