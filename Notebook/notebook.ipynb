# Credit Card Fraud Detection

### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Dataset

```python
PATH = str("../Data/creditcard.csv")
ds = pd.read_csv(PATH)
ds.head()
```

```python
ds.describe()
```

Cheking if dataset has missing values

```python
ds.isnull()
```

Seprate dataset to fraud and none fraud transactions

```python
fraud = ds[ds["Class"] == 1]
valid = ds[ds["Class"] == 0]

print("Valid transaction: ", len(valid))
print("Fraud transaction: ", len(fraud))
print((len(fraud) / len(valid)) * 100)
```

About 0.17% of this dataset includes fraud transactions, so the dataset is crucialy imbalanced

```python
fraud.Amount.describe()
```

```python
valid.Amount.describe()
```

Fraud transaction have higher mean value

### Correlation matrix

```python
correlation_matrix = ds.corr()
sns.heatmap(correlation_matrix, vmax=0.8, square=True)
```

### Data preprocessing

```python
X = ds.iloc[:, :-1].values
y = ds.iloc[:, -1].values

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=32
)
```

### Training Random Forest Classifier Model on the training set

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
yPred = classifier.predict(X_test)
```

### Evaluating the Model

```python
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    precision_score,
    recall_score,
    matthews_corrcoef,
    confusion_matrix,
)


accuracy = accuracy_score(y_test, yPred)
precision = precision_score(y_test, yPred)
recall = recall_score(y_test, yPred)
f1 = f1_score(y_test, yPred)
mcc = matthews_corrcoef(y_test, yPred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"Matthews Correlation Coefficient: {mcc:.4f}")

cm = confusion_matrix(y_test, yPred)
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    xticklabels=["Normal", "Fraud"],
    yticklabels=["Normal", "Fraud"],
)
plt.title("Confusion Matrix")
plt.show()
```

### Applying k-Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(classifier, X_train, y_train, cv=3)
print("Accuracy {:.2f} %".format(scores.mean() * 100))
print("Standard Deviation {:.2f} %".format(scores.std() * 100))
```

By applying k-fold cross validation we can observe that the model has a high accuracy
